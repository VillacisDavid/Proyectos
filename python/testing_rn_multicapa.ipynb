```python
import NeuralNetwork as nn

from sklearn import datasets
import seaborn as sns
```

```python
rn = nn.rn_multicapa([4,10,4,3])
iris = datasets.load_iris()
```

```python
irisdata = iris.data
iristarget = iris.target
```

```python
x = []
for i in irisdata:
    x.append(i.tolist())
z = []
for i in iristarget:
    if(i==0):
        z.append([1,0,0])
    elif(i==1):
        z.append([0,1,0])
    elif(i==2):
        z.append([0,0,1])
    else: print("error")
```

```python
def toVF(x):
    if(x>0.5):
        return 1
    else:
        return 0
```

```python
print("Valores deseados:")
for a,b in zip(x,z):
    print(f"{a}|{b}")
print("Valores predichos:")
print("x1 x2 | x1 XOR x2")
for s in x:
    rn.predict(s)
    y=[]
    vf=[]
    for n in rn.outputLayer:
        y.append(n.y)
        vf.append(toVF(n.y))
    print(f"{s} | {vf} => y = {y}")
```

```python
# Bucle de Ã©pocas de entrenamiento.
for j in range(4000):    
    for i in range(len(x)):
        rn.train(x[i], z[i])
        rn.predict(x[i])
```
